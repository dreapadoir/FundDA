Get the text of Alice in Wonderland using the request method from the urllib package. The book is converted into a list and decoded using the UTF-8 format.

```python
# import urllib to make HTTP requests for internet resources and random to generate random sets 
import urllib.request
import random

# The URL of a text version of Alice in Wonderland.
book_url = 'https://www.gutenberg.org/files/11/11-0.txt'

# Get the book as a list
book = list(urllib.request.urlopen(book_url))

# Decode the lines and strip line endings.
book = [line.decode('utf-8-sig').strip() for line in book]
```

Create a variable called chars that contains a string with all the letters of the alphabet in lower case and the space character.

```python
chars = 'abcdefghijklmnopqrstuvwxyz '
```

Confirm chars is a string.

```python
type(chars)
```

Create a variable sbook to hold the characters in book as a single string. Create a variable weights that holds the occurrences of each character in chars as a list.

```python
# the slicing statement [26:] is used to ignore the 25 line Gutenberg.com preamble at the start of the text
sbook = ''.join(book[26:]).lower()

# Create the weights - count the occurences of each character in the whole book.
weights = [sbook.count(c) for c in chars]

# Show the weights.
weights
```

Create a nested dictionary containing the weights of each letter based on the preceding two letters. 

```python
# Create the weights.
threeLetterWeights ={b: {c: {d: sbook.count(b + c + d) for d in chars} for c in chars} for b in chars}

# Show the weights.
threeLetterWeights
```

Create a dict threeletterdict to hold the occurrences of each three letter permutation. The for loop creates the new dict where the keys are all possible permutations of the nested keys in threeLetterWeights. The values are their occurrences. 

```python
#identify the most common three letter combination
threeletterdict = {}
for i in chars:
    for j in chars:
        for k in chars:
            name = threeLetterWeights[i][j][k]
            #print(i,j,k,' ',name)
            threeletterdict[i + j + k] = name
            
```

```python
threeletterdict
```

Create a variable to determine the most common combination of letters.

```python
mostCommonCombo = max(threeletterdict, key=threeletterdict.get)
print(mostCommonCombo)
```

Use space and t as the inital letters for the new random string, threes, since they are the start of the most common combination. wt turns threeLetterWeights dict into a list. The for loop appends the weighted randon letter output of the nextc variable to the threes string.

```python
threes = ' t'

# Do the following 998 times.
for i in range(1, 999):
    
    wt = threeLetterWeights[threes[-2]]
    wt = wt[threes[-1]]
    # Turn wt into a list, ordered by chars.
    wt = [wt[c] for c in chars]
    # Randomly pick the next character using those weights.
    nextc = random.choices(chars, weights=wt, k=1)[0]
    # Append the character to twos.
    threes = threes + nextc
```

Explain why the log of zero is undefined.

<hr style="border-top: 1px solid rgb(0, 0, 0);" />

The log of $ 0 $ to 2 implies that there is a number, $ x $, such that


$$
2^{x} = 0
$$


It is impossible to raise two to any power to equal zero. 


If $x = 0$, then $2^{0} = 1$. 

If $x > 0$, then $2^{x} \rightarrow \infty $ as $ x \rightarrow \infty $ and $2^{x} > 1$. 

Finally, if $x < 0$, then 

$$2^{x} = \dfrac{1}{2^{-x}}$$

Since in this case $-x > 0$, we can say that

$$\dfrac{1}{2^{-x}} > 0\\
\Rightarrow 2^{x} > 0
$$

Since $$\{0\}\cup\{x > 0\}\cup\{x < 0\} = \mathbb{R}$$ we can show that for all $x \in \mathbb{R}$,

$$
2^{x} > 0
$$

This implies that $2^{x} \neq 0$ for any value $x \in \mathbb{R}$. Therefore, $\log_{2}{0}$ is not defined.
